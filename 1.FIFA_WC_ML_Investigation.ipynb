```python
import pandas as pd
from string import ascii_uppercase as alphabet
import pickle
```

```python
todas_tablas = pd.read_html('https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')
```

# **Extracción de todos los Grupos y convirtiendolos en un archivo pickle**

```python
todas_tablas[9]
todas_tablas[16]
todas_tablas[23]
todas_tablas[30]
todas_tablas[37]
todas_tablas[44]
todas_tablas[51]
todas_tablas[58]
```

```python
alphabet
```

```python
for word, i in zip(alphabet, range(9, 65, 7)):
  print(word, i)
```

```python
dict_tablas = {}
for word, i in zip(alphabet, range(9, 65, 7)):
  df = todas_tablas[i]
  df.rename(columns={df.columns[1]:'Team'}, inplace=True)

  if 'Qualification' in df.columns:
    df.pop('Qualification')
  dict_tablas[f'Grupo {word}'] = df
```
